```
import numpy as np
m=np.array([[1,1,0],[0,2,-1],[2,3,1],[1,0,4]])
m
```

```
n=len(m)
r=np.zeros((n,n))
for i in range(n):
  for j in range(n):
    r[i][j]=sum(abs(m[i]-m[j]))
r
```

```
sum(r)
```
